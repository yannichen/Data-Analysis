## 6. describe the kind of buildings that were sold

```python
import matplotlib.pyplot as plt
#create a dictionary to store value
kind={'one family':0,'office buildings':0,'condo':0, 'other': 0}
for category in data['BUILDING CLASS CATEGORY']:
    if 'ONE FAMILY' in category :
        kind['one family']+=1
    elif 'OFFICE BUILDINGS' in category:
        kind['office buildings']+=1
    if 'CONDO' in category:
        kind['condo']+=1
    else:
        kind['other']+=1
print(kind)
categ = list(kind.keys())
values = list(kind.values())
plt.bar(categ, values, width = 0.4)
plt.show()
```

What I did: I create a new dictionary using type name as key and loop over to count each type, adding to the value corresponding to each type. 

## 7.calculate summary statistics for the prices of properties sold for all of Manhattan and for a couple of select neighborhoods

```python
price=data[data['SALE PRICE']>0]['SALE PRICE']

print("mean:", price.mean())
print("median:", price.median())
print("mode:", price.mode())
print("max:" ,price.max())
print("min:", price.min())
print("25%:", price.quantile(0.25))
print("75%:",price.quantile(0.75))
```

What I did: I removed all 0 values because those are meaningless for analyzing sales. 

```python
#pick two neighborhoods
#extract neighborhood with sale price
price=data[["NEIGHBORHOOD","SALE PRICE"]]
price=price[price['SALE PRICE']>0]
midtowneast=price[price['NEIGHBORHOOD']=='MIDTOWN EAST']
chelsea=price[price['NEIGHBORHOOD']=='CHELSEA']
print('stats for Midtown East:')
print("mean:", midtowneast['SALE PRICE'].mean())
print("median:",midtowneast['SALE PRICE'].median())
print("mode:", midtowneast['SALE PRICE'].mode())
print("max:", midtowneast['SALE PRICE'].max())
print("min:",midtowneast['SALE PRICE'].min())
print("25%:",midtowneast['SALE PRICE'].quantile(0.25))
print("75%:",midtowneast['SALE PRICE'].quantile(0.75))
print()
print('stats for Chelsea:')
print("mean:", chelsea['SALE PRICE'].mean())
print("median:",chelsea['SALE PRICE'].median())
print("mode:", chelsea['SALE PRICE'].mode())
print("max:" ,chelsea['SALE PRICE'].max())
print("min:",chelsea['SALE PRICE'].min())
print("25%:",chelsea['SALE PRICE'].quantile(0.25))
print("75%:",chelsea['SALE PRICE'].quantile(0.75))
```

Though the min for the three situation are all 10, midtown east generally has a lower than the average house in price, because its median, mean 25% and 75% percentiles are all lower than all neighborhoods combined in Manhattan. ## 9. create a visualization the shows the relationship (if any) between the price sold and the number of square footage of all the areas of a property within a building or a structure

```python
from sklearn.linear_model import LinearRegression
relat = data[['GROSS SQUARE FEET','SALE PRICE']].dropna()
relat=relat[relat['SALE PRICE']>10000]
relat=relat[relat['GROSS SQUARE FEET']!=0]
#remove other outliers for analyzing the data set
IQR=relat['SALE PRICE'].quantile(0.75)-relat['SALE PRICE'].quantile(0.25)
upper=relat['SALE PRICE'].quantile(0.75)
lower=relat['SALE PRICE'].quantile(0.25)
relat=relat[relat['SALE PRICE'].between(lower-1.5*IQR,upper+1.5*IQR)]
IQR=relat['GROSS SQUARE FEET'].quantile(0.75)-relat['GROSS SQUARE FEET'].quantile(0.25)
upper=relat['GROSS SQUARE FEET'].quantile(0.75)
lower=relat['GROSS SQUARE FEET'].quantile(0.25)
relat=relat[relat['GROSS SQUARE FEET'].between(lower-1.5*IQR,upper+1.5*IQR)]

#Reshape your data either using array.reshape(-1, 1) if your data has a single feature or array.reshape(1, -1) if it contains a single sample.
x=relat['GROSS SQUARE FEET'].values.reshape(-1, 1)
y=relat['SALE PRICE'].values.reshape(-1, 1)
#plot a best fit line 
bestfit=  LinearRegression()
bestfit.fit(x, y)
y_ =bestfit.predict(x)
plt.scatter(x, y)
plt.ylabel("sale price(in 1e6)")
plt.xlabel("gross area")
plt.plot(x, y_, color="black")
plt.show()
```

what I did: remove all possible outliers using IQR and remove all sales price lower than 10,000; also, I dropped all nan values

we therefore can see a positive correlation between gross square feet and sale price, that when the gross area can be used within a building increases, the sale price will increases as well. 

## 10. which month did the least amount of sales occur

```python
#create a new dictionary to store count values
month = {}
#month 1-12
for i in range(1,13):
    month[i] = 0
#loop over to extract the month from date infromation 
for date in data['SALE DATE']:
    #determine month is 1 or 10-12:
    if date[0] == "1":
        if date[1] =="/":
            month[1] +=1
        else:
            month[int(date[:2])]+=1
    #left are all single number month
    else:
        month[int(date[0])]+=1
```

```python
#visualize the work
categ = list(month.keys())
values = list(month.values())
plt.bar(categ, values, width = 0.4)
plt.xlabel('sale month')
plt.show()
```

What I did: I loop over to extract month information from date and add the value count to the new dictionary I created, with keys as the month number

##### from the visualization we can see that August has the least sell.
##### property sales reaches a lull after June each year, where June is the selling peak.
